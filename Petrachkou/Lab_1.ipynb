Среднее значение, медиана:

```python
def mean(array):
    return Decimal(sum(array)/len(array))

def mediana(array):
    new_array = sorted(array)
    length = len(new_array)
    if (length % 2 == 0):
        return (new_array[length//2] + new_array[length//2 - 1]) / 2
    return new_array[(length-1)//2]
```

Дисперсия, Стандартное отклонение:

```python
def sample_dispertion(array):
    ave = mean(array)
    summa = Decimal(0.0)
    for cell in array:
        summa += ((Decimal(cell) - ave)**2)
    return Decimal(summa/(len(array)-1))

def sigma(array):
    return sample_dispertion(array).sqrt()
```

Квантили уровней 0.01, 0.05, 0.95, 0.99:

```python
def quantil(array, proc):
    new_array = sorted(array)
    return new_array[int(np.floor((len(new_array)-1) * proc))]
```

Вычисление парных корреляций Пирсона:

```python
def PCC(array_1, array_2):
    mean_1 = mean(array_1)
    mean_2 = mean(array_2)
    multi = array_1 * array_2
    cov = mean(multi) - mean_1 * mean_2
    return cov / (sigma(array_1) * sigma(array_2))
```

Представление результатов в виде таблицы:

```python
from decimal import Decimal
size = len(df.columns) + 1
corr_matrix = np.zeros((size, size), dtype=object)
for i in range(1, size):
    corr_matrix[0][i] = df.columns[i-1]
for i in range(1, size):
    corr_matrix[i][0] = df.columns[i-1]

for i in range(1, size):
    for j in range(1, size):
        mean_i = df[df.columns[i-1]].mean()
        mean_j = df[df.columns[j-1]].mean()
        corr_matrix[i][j] = round(PCC(df[df.columns[i-1]].fillna(mean_i), df[df.columns[j-1]].fillna(mean_j)), 2)

print (corr_matrix)
```

Представление результатов в виде DataFrame:

```python
columns = df.columns.tolist()
corr_df = pd.DataFrame(index=columns, columns=columns, dtype=object)
for i in columns:
    for j in columns:
        x = df[i].fillna(df[i].mean())
        y = df[j].fillna(df[j].mean())
        corr_df.loc[i, j] = round(PCC(x, y), 2)
print (corr_df)
```

Как видно, коэффициенты коррреляции рассчитаны верно, так в результате получаем симметричную матрицу с элементами на главной диагонали, равными 1.
При этом, все элементы матрицы не превосходят по модулю значение 1.

График рассеивания:

```python
columns = corr_df.columns.tolist()
for i in range(1, len(columns)):
    for j in range(0, i):
        if ((corr_df.iloc[i, j] >= 0.8) | (corr_df.iloc[i, j] <= -0.8)):
            df.plot.scatter(x=columns[i], y=columns[j], color='red')
            plt.title(f'График рассеивания {columns[i]} vs {columns[j]}')
            plt.xlabel(columns[i])
            plt.ylabel(columns[j])
            plt.grid(True)
            plt.tight_layout()
            plt.show()
```
